## Enamine dose response data exported from CDD
Remove columns and convert csv to tsv

```python
import pandas as pd

dose_df = pd.read_csv('CDD CSV Export - Enamine 68 all 3 runs - 2021-07-21 11h39m09s.csv', keep_default_na=False)
assert dose_df.shape == (88, 14)
assert dose_df['Molecule Name'].unique().size == 68
dose_df.head()
```

```python
# Remove the mostly empty CDD number column and run description column
dose_df = dose_df.drop(labels=['CDD Number', 'PriA-SSB Dose response: Run Conditions'], axis='columns')
assert dose_df.shape == (88, 12)
dose_df.head()
```

```python
# Write to the v1 dataset directory
dose_df.to_csv('../v1/enamine_dose_reponse_curves.tsv', sep='\t', index=False)
```

## Enamine screening data exported from CDD
Convert csv to tsv, remove irrelevant columns, and combine files

```python
def csv_to_tsv(csv_file, tsv_file, cols_to_remove=None):
    csv_df = pd.read_csv(csv_file, keep_default_na=False)
    if cols_to_remove is not None:
        print(f'Original shape: {csv_df.shape}')
        print(f'Remove {len(cols_to_remove)} columns')
        csv_df = csv_df.drop(labels=cols_to_remove, axis='columns')
        print(f'Updated shape: {csv_df.shape}\n')
    csv_df.to_csv(tsv_file, sep='\t', index=False)
```

```python
cols_to_remove = ["IC50 (uM)", "IC50 CI (Lower) (uM)", "IC50 CI (Upper) (uM)",
                  "IC50 Fold Change", "Hill slope", "Span (%)", "Baseline response (%)",
                  "Maximum response (%)", "R squared", "N", "Curve class", "Minimum measured (%)",
                  "Maximum measured (%)", "Concentration (uM) Z'-factor"]

csv_to_tsv('CDD CSV Export - 2021-05-21 Run of PriA-SSB Dose response - exported 2021-08-24.csv',
          'enamine_PriA-SSB_dose_response_2021-05-21_run.tsv', cols_to_remove)
csv_to_tsv('CDD CSV Export - 2021-06-16 Run of PriA-SSB Dose response - exported 2021-08-24.csv',
          'enamine_PriA-SSB_dose_response_2021-06-16_run.tsv', cols_to_remove)
csv_to_tsv('CDD CSV Export - 2021-07-01 Run of PriA-SSB Dose response - exported 2021-08-24.csv',
          'enamine_PriA-SSB_dose_response_2021-07-01_run.tsv', cols_to_remove)
```

```python
# See https://stackoverflow.com/questions/2032403/how-to-create-full-compressed-tar-file-using-python
import tarfile
import glob
enamine_files = glob.glob('enamine_PriA-SSB_dose_response_2021-*_run.tsv')
with tarfile.open('../v1/enamine_PriA-SSB_dose_response_data.tar.gz', "w:gz") as tar:
    for enamine_file in enamine_files:
        tar.add(enamine_file)
```
